```python
import pandas as pd
import numpy as np
import re
import nltk

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

nltk.download('stopwords')
```

```python
import pandas as pd

df = pd.read_csv(
    "fake_reviews.csv",
    engine="python",
    on_bad_lines="skip",
    encoding="utf-8"
)

print(df.shape)
```

```python
print(df.columns)
```

```python
df['text'] = df['review_headline'].fillna('') + " " + df['review_body'].fillna('')
```

```python
df['clean_text'] = df['text'].astype(str)
```

```python
X = df['clean_text']
y = df['fake_review']
```

```python
print("X size:", len(X))
print("y size:", len(y))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=1500,
    stop_words='english'
)

X_vec = tfidf.fit_transform(X)

print("Vector shape:", X_vec.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_vec,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
def predict_review(review):
    review = review.lower()
    vec = tfidf.transform([review])
    result = model.predict(vec)[0]
    return "Fake Review" if result == 1 else "Genuine Review"

print(predict_review("This product is amazing, I love it!"))
print(predict_review("Worst product ever, waste of money"))
```
